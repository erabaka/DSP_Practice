```python
#import class as object
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import style
```

```python
#Defining analog signal

t = np.arange(0,11,0.2)
x = np.sin(t-2)
```

```python
#Plotting analog signal

plt.figure(figsize=(8,6))
plt.suptitle('Analog Signal')
style.use('seaborn-deep')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.plot(t, x, linewidth=3, label='x(t) = sin(t-2)')
plt.xlabel('t', fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.axis([0,10,0,1])
plt.legend(fontsize = 20)
plt.axis([-0.1, 10.1,0,1.5])
plt.xticks(np.arange(0, 16, 1))
plt.yticks(np.arange(-1.2, 1.2, .2))


plt.show()
```

```python
#Sampling

plt.figure(figsize=(8,6))
plt.suptitle('Sampling')
style.use('seaborn-deep')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

n=t
markline, stemlines, baseline = plt.stem(n,x)
plt.setp(stemlines, 'linewidth', 2)


plt.plot(t, x, linewidth=3, label='x(t) = sin(t-2)')
plt.xlabel('t', fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.axis([0,10,0,1])
plt.legend(fontsize = 20)
plt.axis([-0.1, 10.1,0,1.5])
plt.xticks(np.arange(0, 16, 1))
plt.yticks(np.arange(-1.2, 1.2, .2))


plt.show()
```
