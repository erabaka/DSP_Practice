```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

**Generation and Plotting X1 Signal**

```python
x1 = np.array([1,2,2,1,1])
n1 = np.array([0,1,2,3,4])
plt.figure(figsize=(10,8))
style.use('seaborn-deep')

plt.subplot(2,2,1)
markerline, stemlines, baseline = plt.stem(n1,x1,label='X1')
plt.setp(stemlines)
plt.legend(fontsize=20)

# Generation of X2 and plotting

x2 = np.array([3,2,1])
n2 = np.array([0,1,2])

plt.subplot(2,2,2)
markerline, stemlines, baseline = plt.stem(n2,x2,label='X2')
plt.setp(stemlines)
plt.legend(fontsize=20)

#Plotting X1 in position 3

plt.subplot(2,2,3)
markerline, stemlines, baseline = plt.stem(n1,x1,label='X1')
plt.setp(stemlines)
plt.legend(fontsize=20)

#Flip/Fold X2 and plotting

f_x2 = x2[::-1]
n3 = np.array([-2,-1,0])
plt.subplot(2,2,4)
markerline, stemlines, baseline = plt.stem(n3,f_x2,label='Flipped x2')
plt.setp(stemlines)
plt.legend(fontsize=20)

plt.show()
```

Applications of Convolution


```python
# Signal Denoising

#Create signal

Srate = 256 #Hz
t = np.arange(0,3,1/Srate)
pnts = len(t)
print(pnts)
print(t)
```

```python
# Creating a noiseless signal

x = np.sin(2*np.pi * 2 * t)

#Creating a random noise

noise = 5*np.random.randn(pnts)

#Adding noise to signal

NoisySignal = x + noise #Operator overloading, not same as normal addition. More likely, they turn into a similar type and then get added up to position by position
```

```python
# Plotting noisy signal [Pseudo-Continuous]

plt.figure(figsize=(10,8))
style.use('seaborn-deep')
plt.plot(t, NoisySignal, label='Noisy Signal')
plt.legend(fontsize=20)

plt.show()

#plt.show gives more discrete preview
```

```python
#stem makes more analogous view. adding the amplitudes

plt.figure(figsize=(10,8))
style.use('seaborn-deep')
plt.plot(t, NoisySignal, label='Noisy Signal')
plt.legend(fontsize=20)
plt.stem(NoisySignal)
```

```python
# Selecting filters for denoising
filter = 2*np.ones(50)/10.0 #moving avg filter....filter is impulse
print(filter)
```

```python
#Filtering by convolution..
filteredSignal = np.convolve(NoisySignal, filter, mode = 'full')
```

```python
#Plotting filtered signal

plt.figure(figsize=(16,10))
plt.plot(filteredSignal, label='Filtered Signal')
plt.legend(fontsize=20)
plt.show()
```
