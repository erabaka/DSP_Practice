```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

## Answer of Q1

```python
# i)
style.use('seaborn-deep')
# x(n)
x = np.array([3,2,1,1,2,3])
n = np.array([-4,-3,-2,-1,0,1])
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
markerline, stemlines, baseline = plt.stem(n,x,label="signal x(n)")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)

# w(n)
w = np.array([0,0,1.5,0.75,1.5,0])
plt.subplot(1,3,2)
markerline, stemlines, baseline = plt.stem(n,w, label="signal w(n)")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)

# y(n)
# y(n) = 2x(n-1)+w(2n)

x_shift = np.array([3,3,2,1,1,2]) #x(n-1) right shift
x_final = 2*x_shift
w_scale = np.array([0,0,0,1.5,1.5,0])
y = x_final + w_scale
plt.subplot(1,3,3)
markerline, stemlines, baseline = plt.stem(n, y, label='signal y(n)')
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)


# ii)
# Auto Correlation

corr_signal = np.correlate(x, y, mode='same')
plt.figure(figsize=(20,5))
markerline, stemlines, baseline = plt.stem(n, corr_signal, label='correlated signal')
plt.legend(fontsize=12)
plt.show()
```

## Answer to Q2

```python
style.use('seaborn-deep')

# x(n)
x = np.array([2.5,1.5,2,0,2.5])
n = np.array([-2,-1,0,1,2])

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
markerline, stemlines, baseline = plt.stem(n, x, label = "signal x(n)")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)


# w(n)
w = np.array([0,0.5,1.0,0.5,0.8])
plt.subplot(2,2,2)
markerline, stemlines, baseline = plt.stem(n, w, label = "signal w(n)")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)


# noise
noise = x+w
plt.subplot(2,2,3)
markerline, stemlines, baseline = plt.stem(n, noise, label = "noisy signal (x+w)")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)

# filter
filter = np.array([0.0,0.5,0.5,0.5,0.0])
plt.subplot(2,2,4)
markerline, stemlines, baseline = plt.stem(n, filter, label = "filter signal")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=10)


# Convolution to denoise

yc = np.convolve(noise, filter, mode='same')
plt.figure(figsize=(20,5))
markerline, stemlines, baseline = plt.stem(n, yc, label = "denoised signal")
plt.setp(stemlines, 'linewidth', 2)
plt.legend(fontsize=12)

plt.show()
```
