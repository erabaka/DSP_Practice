```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

# **1**. Separate and sketch the complex and real parts of a signal.

```python
t  = np.linspace(0,10,100)  # time
f = 25;                             # frequency
A = 1;                            # amplitude
theta = np.pi/2;               # phase

Complex_wave = A * np.exp(2 + 2j * (np.pi * f * t * theta))

style.use('seaborn-dark')
plt.figure(figsize = (10,8))
plt.plot(t,np.real(Complex_wave),linewidth = 3, label='real part')
plt.plot(t,np.imag(Complex_wave),linewidth = 3, label='imaginary part')
plt.xlim([0,12])
plt.xlabel('time',fontsize = 12)
plt.ylabel('amplitude', fontsize = 12)
plt.title('Complex wave',fontsize = 15)
plt.legend(fontsize = 12)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.plot(t,np.real(Complex_wave),linewidth = 3, label='real part')
plt.xlim([0,12])
plt.xlabel('time',fontsize = 12)
plt.ylabel('amplitude', fontsize = 12)
plt.title('Complex wave Real',fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.plot(t,np.imag(Complex_wave),linewidth = 3, label='imaginary part')
plt.xlim([0,12])
plt.xlabel('time',fontsize = 12)
plt.ylabel('amplitude', fontsize = 12)
plt.title('Complex wave Imaginary',fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
```

```python
# Exponentially reducable signal

t  = np.linspace(0,10,100)  # time
f = 25;                             # frequency
A = 1;                            # amplitude
theta = np.pi/2;               # phase

Complex_wave = np.exp(-0.3 * A * t) * np.exp(2 + 2j * (np.pi * f * t * theta))

style.use('seaborn-dark')
plt.figure(figsize = (10,8))
plt.plot(t,np.real(Complex_wave),linewidth = 3, label='real part')
plt.plot(t,np.imag(Complex_wave),linewidth = 3, label='imaginary part')
plt.xlim([0,12])
plt.xlabel('time',fontsize = 12)
plt.ylabel('amplitude', fontsize = 12)
plt.title('Complex wave',fontsize = 15)
plt.legend(fontsize = 12)
plt.show()
```

2. Perform Z-Transformation and plot the zeros and poles of a given signal.

```python
r = 5
z=r * np.exp(1j * (np.pi/2))
a=0.5
t=np.arange(0,100,0.01)
signal_time_domain = np.sin(t) * (a**t)
plt.plot(t, signal_time_domain)
plt.show()
signal_z_domain = signal_time_domain * z**(-t)
#zero at z=0 and pole at z=a as z/z-a and denominator z-a = 0
zerox=[0]
zeroy=[0]
plt.plot(zerox, zeroy, marker="o", markersize=10)
plt.plot(np.real(signal_z_domain),np.imag(signal_z_domain))
plt.show()

#a circle with radious 0.5 , sunburst
circle=plt.Circle((0,0),0.5)
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot()
ax.set_facecolor('black') #sunburst
ax.add_patch(circle)
ax.axis('equal')
img = plt.show()


```
